```python
import pyttsx3
import speech_recognition as sr
import nltk
```

```python
spEng = pyttsx3.init()
spEng.setProperty('rate', 20)              #Voice rate more:slower
```

```python
spEng.say('India is a great country')
spEng.runAndWait()
```

```python
recognizer = sr.Recognizer()
```

```python
with sr.Microphone() as mic:
    print('Say:',end=' ')
    audio = recognizer.listen(mic, timeout=1, phrase_time_limit=5)
    try:
        text = recognizer.recognize_google(audio)
        print(text)
    except Exception as err:
        print('Could not recognize')
```

```python
with sr.Microphone() as mic:
    print('Say:',end=' ')
    audio = recognizer.listen(mic, timeout=1, phrase_time_limit=10)
    try:
        text = recognizer.recognize_google(audio)
        print(text)
        spEng.setProperty('rate', 40) 
        spEng.say(text)
        spEng.runAndWait()
    except Exception as err:
        print('Could not recognize')
```
